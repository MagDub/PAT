```python
pi_t = 2
beta = 2

v_hit_mat = np.zeros((len(range(1,112)),len(range(-5,5+1))))
for trial in range(1,112):
    for ind, vt in enumerate(range(-5,5+1)):
        norm_trial = (trial-56) if trial>56 else trial
        shrink = (56 - norm_trial)/56
        v_hit = vt + pi_t*shrink
        v_hit_mat[trial-1][ind] = v_hit
    
v_hit_map_flat = v_hit_mat.flatten()
v_hit_map_flat.sort()

p_hit_mat = 1 / (1 + np.exp(beta * (0 - v_hit_map_flat)))

fig, ax = plt.subplots()
ax.plot(v_hit_map_flat, p_hit_mat)

plt.show()
```
