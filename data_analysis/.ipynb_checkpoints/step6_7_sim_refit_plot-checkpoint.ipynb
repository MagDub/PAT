```python
import ipynb.fs.defs.functions_new as fct
import pandas as pd
import numpy as np
```

## Model 0 - alpha

```python
# Settings and parameters
run = 0

# for saving
all_users_folder = 'data/all_users/'

# Model specification
mod_info = {}
mod_info['name'] = 'model0_alpha'
mod_info['value_fct'] = fct.rescorla_wagner_1t
mod_info['dec_fct'] = fct.my_softmax_1t
mod_info['param_names'] = ['alpha']

# Simulations
Nsim = 100

# For fitting
param_lower_bounds = [0]
param_upper_bounds = [1]

# Simulate and fit behaviour, and saves to pickle
if run:
    fct.sim_fit_model(mod_info, param_lower_bounds, param_upper_bounds, Nsim, all_users_folder)
    
# Plot
df = pd.read_pickle(all_users_folder + '/sim_refit/' + mod_info['name'] + '.pkl')
fct.plot_param_recov_conf_matrix(df)
fct.plot_param_recov_correlations(df)
```
