```python
import pandas as pd
import pickle
import numpy as np
```

```python
# Load IDs
with open('uniqueIDs.pkl', 'rb') as f:
    uniqueIDs = pickle.load(f)
```

```python
all_users_folder = 'data/all_users/'
```

# Winning model = model2_v0_alpha_beta_pi_t_newB3

```python
model_folder = all_users_folder + 'fit/model2_v0_alpha_beta_pi_t_newB3/'
data_mod = pd.read_pickle(model_folder+'mod_param_fits.pkl')
mod_info = pd.read_pickle(model_folder+'mod_parameters.pkl')

# Display stats
data_mod_num = data_mod.apply(pd.to_numeric)
data_mod_num.drop('ID', axis=1).describe()

# Load data
ev_per_trial = pd.read_pickle(model_folder + 'mod_ev_per_trial.pkl');
p_hit_per_trial = pd.read_pickle(model_folder + 'mod_p_hit_per_trial.pkl');

PEs = pd.read_pickle(model_folder + 'mod_pe_per_trial.pkl');
hits = pd.read_pickle(model_folder + 'mod_isHit_per_trial.pkl');
trialsNos = pd.read_pickle(model_folder + 'mod_trialNo_per_trial.pkl');
fbs = pd.read_pickle(model_folder + 'mod_fbs_per_trial.pkl');

#shrink_PIs = pd.read_pickle(model_folder + 'shrink_PIs.pkl');
#shrink_alphas = pd.read_pickle(model_folder + 'shrink_alphas.pkl');
```

```python
data_mod_num.to_excel(all_users_folder + 'excel_summary/model_param_summary.xlsx')
```

```python
for ID in uniqueIDs:

    # Extract for ID
    ev_singID = ev_per_trial[ev_per_trial['ID']==ID].drop('ID', axis=1, inplace=False).set_index('Cue',inplace=False)
    pe_singID = PEs[PEs['ID']==ID].drop('ID', axis=1, inplace=False).set_index('Cue',inplace=False)
    phit_singID = p_hit_per_trial[p_hit_per_trial['ID']==ID].drop('ID', axis=1, inplace=False).set_index('Cue',inplace=False)
    trials_singID = trialsNos[trialsNos['ID']==ID].drop('ID', axis=1, inplace=False).set_index('Cue',inplace=False)
    fbs_singID = fbs[fbs['ID']==ID].drop('ID', axis=1, inplace=False).set_index('Cue',inplace=False)
    hits_singID = hits[hits['ID']==ID].drop('ID', axis=1, inplace=False).set_index('Cue',inplace=False)

    # convert
    pe_singID = pe_singID.astype(float)

    cue_ascending = []
    trials_ascending = []
    evs_ascending = []
    pes_ascending = []
    phits_ascending = []
    fbs_ascending = []
    hits_ascending = []
    cue_trials_ascending = []

    Ntrials = 112

    for trial in range(1,Ntrials+1):

        # keep only value on specific trial, df for index
        ev_on_trial = ev_singID[trials_singID==trial]
        cue_ev_on_trial = ev_on_trial.dropna(axis=0, how='all').dropna(axis=1)

        # store cue
        cue_ascending.append(cue_ev_on_trial.index[0])

        # store trial
        trials_ascending.append(trial)

        # store cue_trial
        cue_trial = cue_ev_on_trial.columns[0]
        cue_trials_ascending.append(cue_trial)

        # store EV, PE, phit, fb, hit
        pes_ascending.append(pe_singID.values[trials_singID.values==trial][0])
        evs_ascending.append(ev_singID.values[trials_singID.values==trial][0])
        phits_ascending.append(phit_singID.values[trials_singID.values==trial][0])
        fbs_ascending.append(fbs_singID.values[trials_singID.values==trial][0])
        hits_ascending.append(hits_singID.values[trials_singID.values==trial][0])

    data_per_trial = pd.DataFrame(list(zip(trials_ascending, cue_ascending, cue_trials_ascending, hits_ascending, fbs_ascending,
                                 pes_ascending, evs_ascending, phits_ascending)), 
                        columns=['Trial', 'Cue', 'CueTrial', 'Hit', 'FB',
                                 'mod_PE', 'mod_EV', 'mod_hit_probability'])

    data_per_trial.to_excel(all_users_folder + 'excel_summary/userID_' + ID + '.xlsx')
        
```
