# Functions for plotting

```python
def plot_model_behaviour(fbs, isHit, hit_miss_perc, p_hit_model, vt_model, alpha, v0, beta):

    # Timepoints
    N_trials = 28
    N_iter = int(N_trials/window_size)
    timepoints = ['t'+ str(t+1) for t in range(N_iter)]
    xt = (np.arange(len(timepoints))+1)*window_size

    fig, [ax1, ax2] = plt.subplots(2, 1, figsize=(10,5))

    ### Subplot 1

    # Plot fbs
    height_fb = 1.13
    x = [ind+1 for ind, el in enumerate(fbs) if el == 5]
    l_fb_5 = ax1.scatter(x, height_fb * np.ones(len(x)), c='green', alpha=0.9)
    x = [ind+1 for ind, el in enumerate(fbs) if el == 1]
    l_fb_1 = ax1.scatter(x, height_fb * np.ones(len(x)), c='green', alpha=0.1)
    x = [ind+1 for ind, el in enumerate(fbs) if el == -1]
    l_fb_m1 = ax1.scatter(x, height_fb * np.ones(len(x)), c='red', alpha=0.1)
    x = [ind+1 for ind, el in enumerate(fbs) if el == -5]
    l_fb_m5 = ax1.scatter(x, height_fb * np.ones(len(x)), c='red', alpha=0.9)

    # Plot hits (marker)
    height_hit = 1.2
    x_hit = [ind+1 for ind, el in enumerate(isHit) if el == 1]
    l_hits = ax1.scatter(x_hit, height_hit * np.ones(len(x_hit)), c='black', alpha=1, marker='v')

    # Plot hits (line)
    #l_hits_line, = ax1.plot(list(range(1,len(isHit)+1)), isHit)

    # Plot hit perc (running average)
    l_wind_m, = ax1.plot(xt, hit_miss_perc.loc['hit'], '.-', color='darkorange', alpha=1)

    # Plot probability of hits (from model)
    l_p_h_mod, = ax1.plot(list(range(1,len(p_hit_model)+1)), p_hit_model)

    # Legend
    ax1.legend([l_wind_m, l_p_h_mod], ['Hit running average', 'Hit model probability'], frameon=False)

    # Labels etc 
    ax1.set_ylim([0,1.24])

    # Title 
    ax1.set_title(cue)

    ### Subplot 2

    # Plot expected values (from model)
    l_evs, = ax2.plot(list(range(1,len(vt_model)+1)), vt_model)

    # Labels etc 
    ax2.set_ylim([-5,5])
    ax2.set_yticks(np.arange(-5,5,step=2))
    ax2.set_ylabel('Expected values', fontsize = 14)

    ax2.legend([l_evs], ['EV prediction with: alpha=' + str(alpha)+
                         ', v0='+ str(v0)+', beta='+ str(beta)], frameon=False)

    ### General

    # Same characteristics for each ax
    for ax in [ax1,ax2]:

        # Plot vertical grid
        ax.grid(axis="x", alpha=0.1)

        # Plot horizontal line at 0
        ax.plot(np.arange(-1,30), np.zeros(len(np.arange(-1,30))), 'k--', alpha=0.5, linewidth=1.0)

        # x-axis (trials)
        x_ticks = list(range(1,len(fbs)+1))
        ax.set_xlim(x_ticks[0]-1,x_ticks[-1]+1)
        ax.set_xticks(x_ticks)
        ax.set_xlabel('', fontsize = 18)

    # Display
    plt.show()
```

```python
def compute_2d_values(alpha_values, v0_values, beta):

    # initialise matrix of nlls (+ alphas and v0 for plotting)
    mat_nLL = np.full([len(alpha_values), len(v0_values)], np.nan)
    mat_alpha = np.full([len(alpha_values), len(v0_values)], np.nan)
    mat_v0 = np.full([len(alpha_values), len(v0_values)], np.nan)

    # Compute nLL for different alphas and nLLs
    for i_alpha, alpha in enumerate(alpha_values):
        for i_v0, v0 in enumerate(v0_values):
            mat_nLL[i_alpha][i_v0] = mod1.compute_nLL([v0, alpha, beta])       
            mat_alpha[i_alpha][i_v0] = alpha
            mat_v0[i_alpha][i_v0] = v0
            
    return mat_alpha, mat_v0, mat_nLL
```

```python
def plot_surf_2d(ax, X, Y, Z, beta):
    
    # Plot the surface.
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False, vmin=60, vmax=160)
    
    # Add a color bar which maps values to colors.
    ax.set_zlim(60, 160)
    fig.colorbar(surf, shrink=0.5, aspect=20, ax=ax, location='left') 

    # Labels
    ax.set_zlabel('Negative log likelihood', fontsize = 16)
    ax.set_ylabel('V0', fontsize = 16)
    ax.set_xlabel('alpha (learning rate)', fontsize = 16)
    ax.set_title('Beta (inverse temperature) = ' + str(beta))

    # Extract best parameter
    nLL_min = np.amin(mat_nLL)
    ind_min = np.where(mat_nLL == nLL_min)
    best_alpha = mat_alpha[ind_min]
    best_v0 = mat_v0[ind_min]
    
    print('Best alpha is = ' + str(best_alpha[0]) + ', best v0 is = ' + str(best_v0[0]) + ', min nLL is =' + str(nLL_min))
```

# Minimise nLL with optimisation method (3D): sklearn

```python
ID = '007'

# Get data
user_folder = 'data/user_' + ID + '/'
df2_cf = pd.read_pickle(user_folder + 'df2_cf.pkl')

# Exctract hits and fbs per cue
isHit_all_cues, fbs_all_cues = fct.extract_hits_fbs(df2_cf)

# Create a new Model object
mod1 = fct.Model(mod_name = 'mod1',
                 value_fct = fct.rescorla_wagner, 
                 dec_fct = fct.my_softmax, 
                 param_names = ['v0', 'alpha', 'beta'])

# Input data to model
mod1.set_data(ID, fbs_all_cues, isHit_all_cues)
```

### Initial guess

```python
# ['v0', 'alpha', 'beta']
# define range for input
param_lower_bound = [-5, 0, 0]
param_upper_bound = [ 5, 1, 2]

# define the starting point as a random sample from the domain
initial_guess = np.array(param_lower_bound) + np.random.rand(len(param_lower_bound)) * (np.array(param_upper_bound) - np.array(param_lower_bound))

initial_guess
```

### Nelder-mead optimisation (no gradient)

```python
mat_min_nLL=[]
mat_best_params=[]

for i in range(0,5):

    # define the starting point as a random sample from the domain
    initial_guess = np.array(param_lower_bound) + np.random.rand(len(param_lower_bound)) * (np.array(param_upper_bound) - np.array(param_lower_bound))

    # find the min likelihood
    result = minimize(mod1.compute_nLL, initial_guess, method='nelder-mead',
                   options={'xatol': 1e-8, 'disp': False})

    # store min_nLL and parameters
    mat_min_nLL.append(result.fun)
    mat_best_params.append(result.x)
```

```python
ind = np.argmin(mat_min_nLL)
print(mat_min_nLL[ind])
print(mod1.param_names)
print(mat_best_params[ind])
```

### Powell optimisation

```python
mat_min_nLL=[]
mat_best_params=[]

for i in range(0,5):
    
    # define the starting point as a random sample from the domain
    initial_guess = np.array(param_lower_bound) + np.random.rand(len(param_lower_bound)) * (np.array(param_upper_bound) - np.array(param_lower_bound))
    
    # find the min likelihood
    result = minimize(mod1.compute_nLL, initial_guess, method='Powell',
                   options={'xtol': 1e-8, 'disp': False})

    # store min_nLL and parameters
    mat_min_nLL.append(result.fun)
    mat_best_params.append(result.x)
```

```python
ind = np.argmin(mat_min_nLL)
print(mat_min_nLL[ind])
print(mod1.param_names)
print(mat_best_params[ind])
```

### CG optimisation

```python
mat_min_nLL=[]
mat_best_params=[]

for i in range(0,5):
    
    # define the starting point as a random sample from the domain
    initial_guess = np.array(param_lower_bound) + np.random.rand(len(param_lower_bound)) * (np.array(param_upper_bound) - np.array(param_lower_bound))
    
    # find the min likelihood
    result = minimize(mod1.compute_nLL, initial_guess, method='CG',
                   options={'gtol': 1e-8, 'disp': False})
    
    # store min_nLL and parameters
    mat_min_nLL.append(result.fun)
    mat_best_params.append(result.x)
```

```python
ind = np.argmin(mat_min_nLL)
print(mat_min_nLL[ind])
print(mod1.param_names)
print(mat_best_params[ind])
```

### BFGS optimisation

```python
# direct search method (no gradient computation)
result = minimize(mod1.compute_nLL, initial_guess, method='BFGS',
               options={'gtol': 1e-8, 'disp': True})

print(mod1.param_names)
result.x
```
