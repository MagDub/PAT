```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import ipynb.fs.defs.functions as fct
import pickle

import warnings
warnings.filterwarnings("error")
```

```python
# Load IDs
with open('uniqueIDs.pkl', 'rb') as f:
    uniqueIDs = pickle.load(f)
```

# Fit model for one user

```python
# Parameter range for initial guess 
# ['v0', 'alpha', 'beta']
param_lower_bound = [-5, 0, 0]
param_upper_bound = [ 5, 1, 2]

ID = '007'
        
# Get data
user_folder = 'data/user_' + ID + '/'
df2_cf = pd.read_pickle(user_folder + 'df2_cf.pkl')
isHit_all_cues, fbs_all_cues = fct.extract_hits_fbs(df2_cf)

# Create a new Model object
mod1 = fct.Model(mod_name = 'model1',
                 value_fct = fct.rescorla_wagner, 
                 dec_fct = fct.my_softmax, 
                 param_names = ['v0', 'alpha', 'beta'])

# Input data to model
mod1.set_data(ID, fbs_all_cues, isHit_all_cues)

# Fit model
mod1.fit(param_lower_bound, param_upper_bound, n_iterations=2)
```

### Plot

```python
# Path to running average data
window_size = 4

# Plot
fct.plot_model_behaviour_allcues(mod1, window_size)
```

## Fit for each user and save plot

```python
run_ = False

if run_:
    # Parameter range for initial guess 
    # ['v0', 'alpha', 'beta']
    param_lower_bound = [-5, 0, 0]
    param_upper_bound = [ 5, 1, 2]

    for ID in uniqueIDs: 

        # Get data
        user_folder = 'data/user_' + ID + '/'
        df2_cf = pd.read_pickle(user_folder + 'df2_cf.pkl')
        isHit_all_cues, fbs_all_cues = fct.extract_hits_fbs(df2_cf)

        # Create a new Model object
        mod1 = fct.Model(mod_name = 'model1',
                         value_fct = fct.rescorla_wagner, 
                         dec_fct = fct.my_softmax, 
                         param_names = ['v0', 'alpha', 'beta'])

        # Input data to model
        mod1.set_data(ID, fbs_all_cues, isHit_all_cues)

        # Fit model
        mod1.fit(param_lower_bound, param_upper_bound, n_iterations=2)

        # Path to running average data
        window_size = 4

        # Plot
        fct.plot_model_behaviour_allcues(mod1, window_size, save_fig=True)
```

# Fit mod1 for each participant

```python
run_ = False

if run_:
    
    # Parameter range for initial guess 
    # ['v0', 'alpha', 'beta']
    param_lower_bound = [-5, 0, 0]
    param_upper_bound = [ 5, 1, 2]

    # Fit
    all_users = {}
    for n_part,ID in enumerate(uniqueIDs): 

        # Get data
        user_folder = 'data/user_' + ID + '/'
        df2_cf = pd.read_pickle(user_folder + 'df2_cf.pkl')
        isHit_all_cues, fbs_all_cues = fct.extract_hits_fbs(df2_cf)

        # Create a new Model object
        mod1 = fct.Model(mod_name = 'model1',
                     value_fct = fct.rescorla_wagner, 
                     dec_fct = fct.my_softmax, 
                     param_names = ['v0', 'alpha', 'beta'])

        # Input data to model
        mod1.set_data(ID, fbs_all_cues, isHit_all_cues)

        # Fit model
        mod1.fit(param_lower_bound, param_upper_bound, n_iterations=2)

        # Nested dictionnary user data
        all_users[n_part] = {}
        all_users[n_part]['ID']=mod1.ID
        all_users[n_part]['nLL']=mod1.nLL
        for i in range(0,len(mod1.param_names)):
            all_users[n_part][mod1.param_names[i]]=mod1.param_values[i]

    # Final df
    mod1_fit = pd.DataFrame(all_users).transpose() 
    
    # Save
    all_users_file = 'data/all_users/model_fits/mod1.pkl'
    mod1_fit.to_pickle(all_users_file)
```
