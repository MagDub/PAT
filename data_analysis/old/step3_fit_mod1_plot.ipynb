```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import ipynb.fs.defs.functions as fct
import pickle

import warnings
warnings.filterwarnings("error")
```

# Fit mod1 for each participant

```python
window_size = 16

# Load exclusions
all_users_folder = 'data/all_users/mod1/'
file_name = all_users_folder+'mod_exclusions.pkl'
with open(file_name, "rb") as fp:   
    exclusions = pickle.load(fp)
    
# Load hit_perc_per_t (data per timepoit) and compute stats
all_users_folder = './data/all_users/hit_perc_per_t/'
hit_perc_per_t = pd.read_pickle(all_users_folder + 'hit_perc_per_t_w' + str(window_size) + '.pkl');
hit_perc_per_t = hit_perc_per_t[~hit_perc_per_t['ID'].isin(exclusions)] # rm exclusions
hit_perc_per_t.drop('ID', axis=1, inplace=True);
stats_hits_perc_per_t = hit_perc_per_t.groupby('Code').agg(['mean', 'var']).T.swaplevel(axis=0);

# Load model predictions and compute stats
all_users_folder = './data/all_users/mod1/'
p_hit_per_trial = pd.read_pickle(all_users_folder + 'mod_p_hit_per_trial.pkl');
p_hit_per_trial = p_hit_per_trial[~p_hit_per_trial['ID'].isin(exclusions)] # rm exclusions
p_hit_per_trial.drop('ID', axis=1, inplace=True);
stats_p_hit_per_trial = p_hit_per_trial.groupby('Cue').agg(['mean', 'var']).T.swaplevel(axis=0)
stats_p_hit_per_trial.columns = 'Cue_'+stats_p_hit_per_trial.columns

# Timepoints
N_trials = 112
N_iter = int(N_trials/window_size)
timepoints = ['t'+ str(t+1) for t in range(N_iter)]

# Plot
f, axs = plt.subplots(4, 1, figsize=(10, 10), dpi=80)
plt.subplots_adjust(hspace = 0.4)

x_behav = np.arange(len(timepoints))*4+4
x_mod = np.arange(len(stats_p_hit_per_trial.loc['mean']))+1

for ax, cue, col, alpha in zip(axs, ['Cue_HP', 'Cue_LP', 'Cue_LR', 'Cue_HR'], ['red', 'red', 'green', 'green'], [0.6, 0.3, 0.3, 0.6]):

    # Behaviour (running average)
    m=ax.plot(x_behav, stats_hits_perc_per_t.loc['mean'][cue], 'o-', color='blue', alpha=0.5);
    vm=ax.plot(x_behav, stats_hits_perc_per_t.loc['mean'][cue]-stats_hits_perc_per_t.loc['var'][cue], ':', color='blue', alpha=0.5)
    vp=ax.plot(x_behav, stats_hits_perc_per_t.loc['mean'][cue]+stats_hits_perc_per_t.loc['var'][cue], ':', color='blue', alpha=0.5)
    
    # Model predictions
    m=ax.plot(x_mod, stats_p_hit_per_trial.loc['mean'][cue], '.-', color=col, alpha=alpha);
    vm=ax.plot(x_mod, stats_p_hit_per_trial.loc['mean'][cue]-stats_p_hit_per_trial.loc['var'][cue], ':', color=col, alpha=alpha)
    vp=ax.plot(x_mod, stats_p_hit_per_trial.loc['mean'][cue]+stats_p_hit_per_trial.loc['var'][cue], ':', color=col, alpha=alpha)

    ax.grid(axis='x', color='0.95');
    ax.set_title(cue, fontsize=16);
    ax.set_xticks(x_mod);
    ax.set_ylabel("Hits [%]", fontsize=16);
    ax.set_ylim((0,1));

plt.show()
```

```python
# Load model predictions and compute stats
all_users_folder = './data/all_users/mod1/'

ev_per_trial = pd.read_pickle(all_users_folder + 'mod_ev_per_trial.pkl');
ev_per_trial = ev_per_trial[~ev_per_trial['ID'].isin(exclusions)] # rm exclusions
ev_per_trial.drop('ID', axis=1, inplace=True);
stats_ev_per_trial = ev_per_trial.groupby('Cue').agg(['mean', 'var']).T.swaplevel(axis=0)
stats_ev_per_trial.columns = 'Cue_'+stats_ev_per_trial.columns

p_hit_per_trial = pd.read_pickle(all_users_folder + 'mod_p_hit_per_trial.pkl');
p_hit_per_trial = p_hit_per_trial[~p_hit_per_trial['ID'].isin(exclusions)] # rm exclusions
p_hit_per_trial.drop('ID', axis=1, inplace=True);
stats_p_hit_per_trial = p_hit_per_trial.groupby('Cue').agg(['mean', 'var']).T.swaplevel(axis=0)
stats_p_hit_per_trial.columns = 'Cue_'+stats_p_hit_per_trial.columns

# Plot
f, axs = plt.subplots(2, 1, figsize=(10, 6), dpi=80)
plt.subplots_adjust(hspace = 0.4)

x_mod = np.arange(len(stats_ev_per_trial.loc['mean']))+1

# Hit probabilites
m1=axs[0].plot(x_mod, stats_p_hit_per_trial.loc['mean']['Cue_HP'], '.-', color='red', alpha=0.6);
m=axs[0].plot(x_mod, stats_p_hit_per_trial.loc['mean']['Cue_LP'], '.-', color='red', alpha=0.3);
m=axs[0].plot(x_mod, stats_p_hit_per_trial.loc['mean']['Cue_LR'], '.-', color='green', alpha=0.3);
m=axs[0].plot(x_mod, stats_p_hit_per_trial.loc['mean']['Cue_HR'], '.-', color='green', alpha=0.6);

# Behaviour: sliding average
b1=axs[0].plot(x_behav, stats_hits_perc_per_t.loc['mean']['Cue_HP'], '.-', color='gray', alpha=0.6);
b=axs[0].plot(x_behav, stats_hits_perc_per_t.loc['mean']['Cue_LP'], '.-', color='gray', alpha=0.3);
b=axs[0].plot(x_behav, stats_hits_perc_per_t.loc['mean']['Cue_HR'], '.-', color='gray', alpha=0.6);
b=axs[0].plot(x_behav, stats_hits_perc_per_t.loc['mean']['Cue_LR'], '.-', color='gray', alpha=0.3);

axs[0].grid(axis='x', color='0.95');
axs[0].set_title('Mod1 probability of hit', fontsize=16);
axs[0].set_xticks(x_mod);
axs[0].set_ylabel("Hit [%]", fontsize=16);
axs[0].legend(b1, {'Behaviour'})

# Expected values
m=axs[1].plot(x_mod, stats_ev_per_trial.loc['mean']['Cue_HP'], '.-', color='red', alpha=0.6);
m=axs[1].plot(x_mod, stats_ev_per_trial.loc['mean']['Cue_LP'], '.-', color='red', alpha=0.3);
m=axs[1].plot(x_mod, stats_ev_per_trial.loc['mean']['Cue_LR'], '.-', color='green', alpha=0.3);
m=axs[1].plot(x_mod, stats_ev_per_trial.loc['mean']['Cue_HR'], '.-', color='green', alpha=0.6);

axs[1].grid(axis='x', color='0.95');
axs[1].set_title('Mod1 expected values', fontsize=16);
axs[1].set_xticks(x_mod);
axs[1].set_ylabel("Values", fontsize=16);

plt.show()
```
