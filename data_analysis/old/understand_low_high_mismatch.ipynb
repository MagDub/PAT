```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Get indiv data
# user_folder = 'data/user_' + ID + '/'
# df2_cf = pd.read_pickle(user_folder + 'df2_cf.pkl')

# Get average data
all_users_folder = 'data/all_users/mod1/'
mod_fit = pd.read_pickle(all_users_folder+'mod_param_fits.pkl')
p_hit_per_trial = pd.read_pickle(all_users_folder+'mod_p_hit_per_trial.pkl')
ev_per_trial = pd.read_pickle(all_users_folder+'mod_ev_per_trial.pkl')
```

```python
indiv_ev_per_trial = ev_per_trial[ev_per_trial['ID']=='007']

indiv_ev_per_trial
```

```python
R_cues = indiv_ev_per_trial.set_index('Cue').T.drop('ID', axis=0).drop('LP', axis=1).drop('HP', axis=1)

R_cues.head()
```

```python
# Plot
f, axs = plt.subplots(2, 1, figsize=(10, 6), dpi=80)
plt.subplots_adjust(hspace = 0.4)

x_mod = np.arange(len(R_cues))+1

for cue, alpha in zip(['LR', 'HR'], [0.3, 0.6]):
    # Hit probabilites
    m=axs[0].plot(x_mod, R_cues[cue], '.-', color='green', alpha=alpha);
    # Expected values
    m=axs[1].plot(x_mod, R_cues[cue], '.-', color='green', alpha=alpha);
```
